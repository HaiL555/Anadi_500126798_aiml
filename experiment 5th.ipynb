```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Function to load data from a CSV file
def load_data(file_path):
    data = pd.read_csv('/content/renewable-energy-stock-account-2007-2022.csv')
    return data

# Function to perform basic exploratory data analysis
def exploratory_data_analysis(data):
    # 1. Display the first five rows of the dataset
    print("First five rows of the dataset:")
    print(data.head(), "\n")

    # 2. Check data types and missing values
    print("Data types of each column:")
    print(data.dtypes, "\n")

    print("Missing values in each column:")
    print(data.isnull().sum(), "\n")

    # 3. Summary statistics of numerical columns
    print("Summary statistics for numerical columns:")
    print(data.describe(), "\n")

    # 4. Summary statistics for categorical columns
    print("Summary statistics for categorical columns:")
    print(data.describe(include=['object']), "\n")

    # 5. Unique values in categorical columns
    print("Unique values in each categorical column:")
    for col in data.select_dtypes(include=['object']):
        print(f"{col}: {data[col].nunique()} unique values")
    print("\n")

    # 6. Visualize the distribution of numerical columns
    print("Distribution plots for numerical columns:")
    numerical_columns = data.select_dtypes(include=['number']).columns
    for col in numerical_columns:
        plt.figure(figsize=(8, 4))
        sns.histplot(data[col].dropna(), kde=True)
        plt.title(f"Distribution of {col}")
        plt.show()

    # 7. Pairplot for visualizing relationships between numerical columns
    print("Pairplot for numerical columns (first 5 numerical columns):")
    sns.pairplot(data[numerical_columns[:5]].dropna())
    plt.show()

    # 8. Correlation heatmap
    print("Correlation heatmap of numerical columns:")
    plt.figure(figsize=(10, 8))
    sns.heatmap(data[numerical_columns].corr(), annot=True, cmap='coolwarm')
    plt.title("Correlation heatmap")
    plt.show()

# Main execution
if __name__ == "__main__":
    # Load dataset from a CSV file
    file_path = 'your_dataset_file.csv'  # Replace with your actual file path
    data = load_data(file_path)

    # Perform exploratory data analysis
    exploratory_data_analysis(data)
```
